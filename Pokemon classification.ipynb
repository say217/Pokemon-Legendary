```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
```

```python
X = new_df.drop(columns=["is_legendary"])
y = new_df["is_legendary"]
```

```python
y
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
X_train
```

```python
pokemon_model = RandomForestRegressor(n_estimators=200, random_state=42)
pokemon_model.fit(X_train, y_train)
```

```python
y_pred = pokemon_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print('The model evaluation complete')
print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
print("R_2:", r2)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

tree_index = 0
tree = pokemon_model.estimators_[tree_index]
plt.figure(figsize=(30, 20))
plot_tree(tree, filled=True, rounded=True, fontsize=10)


plt.title(f'Random Forest Tree #{tree_index + 1}', fontsize=20)

plt.savefig(f'random_forest_tree_{tree_index + 1}.png', bbox_inches='tight')  # saves to current folder
plt.show()

```
